```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import MeCab

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("question.csv")
df = df.fillna(0)
category_name = df.columns[1:]
category_name
```

### テキストを形態素に分解

```python
mecab = MeCab.Tagger('-d /usr/local/lib/mecab/dic/mecab-ipadic-neologd')

def parse_text(text):
    words = []
    node = mecab.parseToNode(text)
    while node:
        word = node.surface
        node = node.next
        words.append(word)
    words = " ".join(words)
    return words

df["words"] = df["自問"].apply(parse_text)
df
```

### 入力データの整形

```python
# 訓練データとテストデータに分割
train_df, test_df = train_test_split(df, test_size=0.3)

# TF-IDFで形態素をベクトル化
vectorizer = TfidfVectorizer(ngram_range=(1,3), stop_words=[","], max_features=3000)
vectorizer.fit(df["words"].values)

def preprocess(df):
    Q = df["自問"].values
    X = vectorizer.transform(df["words"].values).toarray()
    Y = df.iloc[:, 1:10].values
    return Q, X, Y

Q_train, X_train, Y_train = preprocess(train_df)
Q_test, X_test, Y_test = preprocess(test_df)
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

### 分類器を訓練

```python
# 分類器はロジスティック回帰モデル
estimator = LogisticRegression(solver='sag')
# 一対他の分類課題とする。
# 例えば、「メンバー成長支援」というラベルと「その他8つのどれか」というラベルの二値分類モデルを、ラベルと同じ数（=9個）作る
classifier = OneVsRestClassifier(estimator)
# 訓練
classifier.fit(X_train, Y_train)
# テストデータに対する推定結果を取得
Y_prob = classifier.predict_proba(X_test)
```

### 推定結果例

```python
# テストデータのi番目の自問についてラベルごとの推定スコアをプロット
for i in range(len(Y_test)):
    fig = plt.figure(figsize=(6,6))
    plt.barh(category_name, Y_prob[i], label="Prediction")
    plt.barh(category_name, Y_test[i], alpha=0.2, label="Ground Truth")
    plt.title(Q_test[i])
    plt.xlim(0,1)
    plt.xlabel("Predicted Score", fontsize=15)
    plt.grid()
    plt.legend()
    plt.show()
```

### 精度評価
精度　= (分類器が推定したスコアが最も高いカテゴリが教師データに含まれていた自問の数) / (自問の数)

```python
# スコアの最も高いラベルを取得
Y_pred = np.argmax(Y_prob, axis=1)
# 正解に含まれているラベルを取得
Y_gt = [np.where(y == 1)[0] for y in Y_test]
# 精度
accuracy = np.sum([pred in gt for pred, gt in zip(Y_pred, Y_gt)]) / len(Y_gt)
print("精度:", accuracy)
```
